### 4.Visualization
##### 4.1) Visualization <Histplot>


The Histogram data visualization for the Adidas dataset
offers a graphical representation of the frequency
distribution of a specific column's values.

This visualization divides the data into intervals, or
bins, and displays the count of data points falling within
each bin as vertical bars.

The Histogram provides insightsinto the distribution
pattern, central tendencies, and outliers of the
selected column.

By observing the height and shape of the bars, professionals
can understand the concentration of values and their ranges.

This visualization aids in identifying trends and variations in
the data, contributing to data-driven decision-making and actionable
insights for Adidas's business strategies.

```python
plt.figsize=(32,6)
adidas.hist()
plt.show()
```

##### 4.2) Visualization <Boxplot>


The Boxplot data visualization for the Adidas dataset's
columns "Units Sold" and "Retailer ID" offers a visual summary
of the distribution, central tendency, and potential
outliers within these variables.

This graphical representation presents a box-and-whisker
plot that showcases the range, median, quartiles, and
possible extreme values of units sold by different retailer IDs.

By observing the boxplot's elements, analysts can understand the
sales distribution across retailers and identify potential
performance variations or anomalies.

The Boxplot visualization enhances data analysis by offering
insights into sales trends and retailer-specific performance,
guiding decisions related to marketing strategies, sales optimization, and collaboration with retailers for Adidas.

```python
#Checking for outliars
import matplotlib.pyplot as plt
plt.figure(1)
plt.boxplot([adidas['Units Sold']])
plt.figure(2)
plt.boxplot([adidas['Retailer ID']])
plt.show()
```
